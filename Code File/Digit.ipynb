```
import numpy as np
```

```
from sklearn.neural_network import MLPClassifier
```

```
from sklearn.metrics import confusion_matrix
```

```
import mnist as mn
```

```
x_train=mn.train_images()
```

```
y_train=mn.train_labels()
```

```
x_test=mn.test_images()
```

```
y_test=mn.test_labels()
```

```
print(x_train.shape)
print(x_test.shape)
```

```
x_train=x_train.reshape((60000,28*28))
```

```
x_test=x_test.reshape((-1,28*28))
```

```
x_train=x_train/256
x_test=x_test/256
```

```
clf=MLPClassifier( hidden_layer_sizes=(64,64),activation='relu', solver='adam')
```

```
clf.fit(x_train,y_train)
```

```
pred=clf.predict(x_test)
```

```
acc=confusion_matrix(y_test,pred)
acc
```

```
def accuracy(cm):
  dia=cm.trace()
  ele=cm.sum()
  return dia/ele
```

```
print(accuracy(acc))
```
