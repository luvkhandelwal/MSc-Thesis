```python
import pandas as pd
import numpy as np
 
# Loading the dataset
data = pd.read_csv("death_severe_cases.csv")
 
#Using Correlation analysis to depict the relationship between the numeric/continuous data variables
corr = data.corr()
print(corr)
```

```python
print(list(data.columns))
```

```python
#importing library
import matplotlib.pyplot as plt
#scatter plot for the dataset
plt.scatter(data.deaths, data.severe)
plt.show()
```
