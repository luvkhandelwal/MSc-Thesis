```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import statsmodels.api as sm
clean_data=pd.read_csv('complete_dataset_cleaned.csv')
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
data=clean_data.iloc[:,2:]
data=data.iloc[:,:-2]
data.index=data.location
data.pop('date')
data.pop('location')
data.pop('human_development_index')
data.pop('gdp_percentage_weekly')

from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(data)
c=data.columns
i=data.index

df = pd.DataFrame(scaler.transform(data))

df.columns=c
df.index=i
df
```

```python
df=clean_dataset(df)
```

```python
df=df.groupby(df.index).mean()
```

```python
df=df.iloc[:,[0,1,2,3,4,5,13]]
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]

#fit linear regression model
model = sm.OLS(y, X).fit()

#view model summary
print(model.summary())
```

```python
c=list(model.params)
c=pd.DataFrame(c)
c.to_csv('importance.csv') 
```
